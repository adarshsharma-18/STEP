```python
from abc import abstractmethod

class programmingLanaguage:
    @abstractmethod
    def getName(self):
        pass

    @abstractmethod
    def getType(self):
        pass

class procedural(programmingLanaguage):
    def gettype(self):
        return "Procedural"
    
    
class objectOriented(programmingLanaguage):
    def gettype(self):
        return "Object Oriented"
    

class hybProOO(procedural,objectOriented):
    def gettype(self):
        return procedural.gettype(self)  + objectOriented.gettype(self)
    
    
class LangC(procedural):
    def getname(self):
        return "C"

class LangCPP(hybProOO):
    def getname(self):
        return "C++"
class LangJava(objectOriented):
    def getname(self):
        return "java"
class Langpython(hybProOO):
    def getname(self):
        return "python"

java = LangJava()
print(java.gettype())
print(java.getname())

cpp = LangCPP()
print(cpp.gettype())
print(cpp.getname())
```

```python
from abc import abstractmethod

class programmingLanaguage:
    @abstractmethod
    def getName(self):
        pass

    @abstractmethod
    def getType(self):
        pass

class procedural(programmingLanaguage):
    def gettype(self):
        return " Procedural"
    
    
class objectOriented(programmingLanaguage):
    def gettype(self):
        return " Object Oriented"
    
class aspectOriented(programmingLanaguage):
    def gettype(self):
        return " Aspect Oriented"

class functional(programmingLanaguage):
    def gettype(self):
        return " functional"
    

class hybrid1(procedural,objectOriented,aspectOriented):
    def gettype(self):
        return procedural.gettype(self)  + objectOriented.gettype(self) + aspectOriented.gettype(self)
    
class hybrid2(objectOriented,aspectOriented,functional):
    def gettype(self):
        return   objectOriented.gettype(self) + aspectOriented.gettype(self)+ functional.gettype(self)

class hybrid3(procedural,objectOriented,aspectOriented,functional):
    def gettype(self):
        return procedural.gettype(self)  + objectOriented.gettype(self) + aspectOriented.gettype(self) + functional.gettype(self)
    
class LangC(procedural):
    def getname(self):
        return "C"

class LangCPP(hybrid1):
    def getname(self):
        return "C++"
    
class LangJava(hybrid2):
    def getname(self):
        return "java"
    
class Langpython(hybrid3):
    def getname(self):
        return "python"

java = LangJava()
print(java.gettype())
print(java.getname())

cpp = LangCPP()
print(cpp.gettype())
print(cpp.getname())

py = Langpython()
print(py.gettype())
print(py.getname())
```
